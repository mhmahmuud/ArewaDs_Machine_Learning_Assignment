```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey = pd.read_csv('honey.csv')
honey.head()
```

```python
# Arizona
az_honey=honey[honey['state']=='AZ']
az_honey
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x= az_honey['year'], y=az_honey['numcol'], data=az_honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Honey Production Year over Year')

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=az_honey['year'], y=az_honey['yieldpercol'], ax=ax2, color="r", 
                         label ='Yield per colony', legend=False) 
sns.despine(right=False)
plt.ylabel('colony yield')
ax.figure.legend()
```
