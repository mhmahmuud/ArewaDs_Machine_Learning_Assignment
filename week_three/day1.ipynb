```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the data
birds_df = pd.read_csv("birds.csv")
```

```python
#
birds_df.head()
```

```python
wingspan = birds_df[birds_df['MaxBodyMass'] > birds_df['MaxBodyMass'].mean()]
wingspan['MaxBodyMass'].plot()
birds_df['MaxBodyMass'].mean()
```

```python
geese = []
for snow in birds_df['Name']:
    if 'goose' in snow:
        geese.append(snow)
geese
```

```python
geese_df = birds_df[birds_df['Name'].isin(geese)]
```

```python
plt.xlabel("Geese")
plt.ylabel('MaxBodyMass (cm)')
plt.title("Max BodyMass in Centimeters")
plt.plot(geese_df['Name'],geese_df['MaxBodyMass'])
plt.show()
```

```python
plt.xlabel("Geese")
plt.ylabel('MaxBodyMass (cm)')
plt.title("Max BodyMass in Centimeters")
plt.bar(geese_df['Name'],geese_df['MaxBodyMass'])
plt.show()
```

```python
plt.xlabel("Geese")
plt.ylabel('MaxBodyMass (cm)')
plt.title("Max BodyMass in Centimeters")
plt.scatter(geese_df['Name'],geese_df['MaxBodyMass'])
plt.show()
```

from the above 3 plots one could tell that Canada Goose has the highest maximum Body Mass and Ross's Goose has the lowest maximum Body Mass.
