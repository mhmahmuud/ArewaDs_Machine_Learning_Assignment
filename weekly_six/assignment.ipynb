```python
# importing the needed libraries 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# loading data
df = pd.read_csv('../data/nigerian-songs.csv')
df.head()
```

```python
plt.scatter(x=df['loudness'],y=df['energy'])
```

```python
sns.jointplot(data=df, x="popularity", y= "danceability",hue="artist_top_genre", kind='scatter')
```

I dont think Missing values of artist_top_genre column would do any right to our data so is better to drop the missing values.

```python
df = df = df[df['artist_top_genre'] != 'Missing']
```

```python

sns.set_style('white')
sns.jointplot(data=df, x="popularity", y= "danceability",hue="artist_top_genre", kind='scatter')
sns.relplot(data=df, x="popularity", y= "danceability",hue="artist_top_genre", kind='scatter')
sns.despine(left=True)
```

```python
sns.scatterplot(data=df, x="popularity", y= "danceability",hue="artist_top_genre")
```

from the visualization above one could say 
- The mean songs listened to are from nigeria 
- Afroswing has the highest danceability
- Nigerian pop has the lowest danceability


```python
sns.jointplot(data=df, x="energy", y= "danceability",hue="artist_top_genre", kind='scatter')
```

```python
sns.jointplot(data=df, x="energy", y= "loudness",hue="artist_top_genre", kind='scatter')
```
