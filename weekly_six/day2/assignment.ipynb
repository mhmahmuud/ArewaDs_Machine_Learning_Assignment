```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=100, centers=3, random_state=0)

# Perform hierarchical clustering
clustering = AgglomerativeClustering(n_clusters=3)
clustering.fit(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=clustering.labels_, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hierarchical Clustering')
plt.show()
```

In the above example, I first generated a synthetic dataset using make_blobs() function from the sklearn.datasets module,I also create 100 samples with 2 features (X) and assign them to 3 clusters (y).

Next, I initialize an instance of AgglomerativeClustering and specify the number of clusters (n_clusters=3). I then fit the hierarchical clustering model to the data using fit(X).

Finally, I visualize the clusters by plotting the data points, coloring them according to their cluster labels (clustering.labels_), and adding appropriate labels and titles to the plot.


```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# Perform DBSCAN clustering
clustering = DBSCAN(eps=0.3, min_samples=5)
clustering.fit(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=clustering.labels_, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering')
plt.show()
```

In the above example, I first generated a synthetic dataset using make_moons() function from the sklearn.datasets module, I also created 200 samples representing two interleaving half circles (moons) with some added noise.

Next, I initialize an instance of DBSCAN and specify the parameters. ***eps*** determines the maximum distance between two samples for them to be considered as neighbors, and **min_samples** sets the minimum number of samples in a neighborhood for a point to be considered as a core point.

I then fit the DBSCAN model to the data using fit(X). The algorithm automatically identifies core samples, border samples, and noise points.

Finally, I visualize the clusters by plotting the data points, coloring them according to their cluster labels (clustering.labels_), and adding appropriate labels and titles to the plot.
